### Start Time

### question
- what time of the day the accidents mostly happen?
- which day of the week have the most accidents?
- which months have the most accidents?
- what is the trend of the accidents year over year
-

```
data['Start_Time']
```

- we have to convert Start_Time from string to datetime

```
data['Start_Time']=pd.to_datetime(data['Start_Time'])
```

```
data['Hour']=data['Start_Time'].dt.hour
```

```
sns.distplot(data['Hour'],bins=24,kde=False,norm_hist = True)
```

- we can see that from above plot there are high % of accidents happens in the morning time between 7 to 9
- next highest % is in between 3 pm to 6 pm

```
sns.distplot(data['Start_Time'].dt.dayofweek,bins=7,kde=False,norm_hist = True)
```

```
sundays_start_time = data.Start_Time[data.Start_Time.dt.dayofweek == 6]
sns.distplot(sundays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```
monday_start_time = data.Start_Time[data.Start_Time.dt.dayofweek == 0]
sns.distplot(monday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```
data.Source.value_counts().plot(kind='pie')
```

```
data_2019 = data[data.Start_Time.dt.year == 2019]
data_2019_Bing = data_2019[data_2019.Source == 'MapQuest']
sns.distplot(data_2019_Bing.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```
